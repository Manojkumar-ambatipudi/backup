```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ---------- Question 8 ----------


# ---------- Parameters ----------
m = 2
t_min = -np.pi / 3
t_max = np.pi / 3

t = np.linspace(t_min, t_max, 2000)

def p(t, x):
    return x[0] * np.cos(t) + x[1] * np.cos(2 * t)

def in_S(x):
    values = p(t, x)
    return np.max(np.abs(values)) <= 1.0 + 1e-3  

x1 = np.array([0.5, 0.2])
x2 = np.array([-0.4, 0.3])


# ---------- Plot 1: Trigonometric polynomials ----------
p1 = p(t, x1)
p2 = p(t, x2)
p_avg = 0.5 * (p1 + p2)

plt.figure()
plt.plot(t, p1)
plt.plot(t, p2)
plt.plot(t, p_avg, linestyle='--')
plt.axhline(1)
plt.axhline(-1)
plt.title("Trigonometric Polynomials vs t")
plt.xlabel("t")
plt.ylabel("p(t)")
plt.show()

# Create grid in x1, x2
x1_vals = np.linspace(-2, 2, 800)
x2_vals = np.linspace(-2, 2, 800)

X1, X2 = np.meshgrid(x1_vals, x2_vals)
Z = np.zeros_like(X1)

for i in range(X1.shape[0]):
    for j in range(X1.shape[1]):
        x = np.array([X1[i, j], X2[i, j]])
        Z[i, j] = np.max(np.abs(p(t, x)))

# ---------- Plot 2: Approximate boundary of S ----------
plt.figure()
plt.contour(X1, X2, Z, levels=[1])
plt.title("Numerical Approximation of Set S (m=2)")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
# ---------- Question 5 ----------
from mpl_toolkits.mplot3d import Axes3D

# Create grid for x and z (both must be >= 0)
x_vals = np.linspace(0, 3, 2000)
z_vals = np.linspace(0, 3, 2000)
X, Z = np.meshgrid(x_vals, z_vals)

# Boundary surface: determinant = 0  ->  y^2 = xz
Y_pos = np.sqrt(X * Z)
Y_neg = -np.sqrt(X * Z)

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot boundary surface (both sheets)
ax.plot_surface(X, Y_pos, Z, alpha=0.8)
ax.plot_surface(X, Y_neg, Z, alpha=0.8)

# Labels
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.show()
```
